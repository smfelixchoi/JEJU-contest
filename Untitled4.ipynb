```python
df.rename(columns = {'REG_YYMM' : 'DATE', 'CARD_SIDO_NM' :'SIDO', 
                     'STD_CLSS_NM' : 'STD', 'SEX_CTGO_CD':'SEX'}, inplace = True)


df['DATE'] = pd.to_datetime(df['DATE'],format='%Y%m',errors='coerce')
df['year'] = df['DATE'].dt.year
df['month'] = df['DATE'].dt.month

drop_columns = ['DATE', 'CARD_CCG_NM', 'HOM_SIDO_NM', 'HOM_CCG_NM', 'CSTMR_CNT']
df.drop(drop_columns, axis=1, inplace = True)

print(df.shape)
df.head(1)
```

```python
#AMT_mean column 추가
df['AMT_mean'] = np.round(df['AMT'] / df['CNT'])
```

```python
#이상치 제거하는 코드 업데이트!

def no_outliers(df):
    Q1 = df['AMT_mean'].quantile(0.25)
    Q3 = df['AMT_mean'].quantile(0.75)
    IQR = Q3 - Q1
    lower_bound = Q1 - 1.5 * IQR
    upper_bound = Q3 + 1.5 * IQR
    amt_mean_outlier_indices = df[(df['AMT_mean'] < lower_bound) | (df['AMT_mean'] > upper_bound)].index
    
    Q11 = df['CNT'].quantile(0.25)
    Q33 = df['CNT'].quantile(0.75)
    IQRR = Q33 - Q11
    lowerbound = Q11 - 1.5 * IQRR
    upperbound = Q33 + 1.5 * IQRR
    cnt_outlier_indices = df[(df['CNT'] < lowerbound) | (df['CNT'] > upperbound)].index
    
    total_outlier_indices = sorted(list(set(amt_mean_outlier_indices.tolist() + cnt_outlier_indices.tolist())))
    df_out = df.drop(total_outlier_indices, axis = 0, inplace = False)
    
    return df_out
```

```python
print(df.shape)

df = no_outliers(df)
print(df.shape)
```

```python
#위에서 만든 열 'AMT_mean'은 이상치를 제거하기 위해 만들었던 열이므로 다시 지운다. 
#뒤에서 groupby를 수행하면서 값에 오차가 커지기 때문 (np.round 이용했으므로)
df.drop(['AMT_mean'], axis = 1, inplace = True)

df = df.groupby(['year','month', 'SIDO', 'STD', 'AGE', 'SEX', 'FLC'])['AMT','CNT'].sum().reset_index()

#다시 정확한 'AMT_mean'을 만듦
df['AMT_mean'] = np.round(df['AMT'] / df['CNT'])

#'AMT'의 예측은 'AMT_mean'의 예측값과 'CNT'의 예측값의 곱을 통해서 얻는다.
df.drop(['AMT'], axis = 1, inplace = True)

#groupby를 통해 데이터의 수가 많이 줄어든 것을 알 수 있다. 
print(df.shape)
df.head(1)
```

```python
# target 로그 변환
df['CNT'] = np.log1p(df['CNT'])
df['AMT_mean'] = np.log1p(df['AMT_mean'])


# one-hot-encoding을 위해 str으로 바꿈
cols = ['SEX', 'FLC', 'year', 'month']
for col in cols:
    df[col] = df[col].apply(str)
```

```python
def AMT_train_test(df):
    target = df['AMT_mean']
    X_features = df.drop(['AMT_mean','CNT'], axis = 1)
    X_train, X_test, y_train, y_test = train_test_split(X_features, target, test_size = 0.2, random_state = 10)
    return target, X_features, X_train, X_test, y_train, y_test

def CNT_train_test(df):
    target = df['CNT']
    X_features = df.drop(['AMT_mean','CNT'], axis = 1)
    X_train, X_test, y_train, y_test = train_test_split(X_features, target, test_size = 0.2, random_state = 10)
    return target, X_features, X_train, X_test, y_train, y_test
```

```python
def get_top_bottom_coef(model, n=10):
    coef = pd.Series(model.coef_, index = X_features.columns)
    
    coef_high = coef.sort_values(ascending = False).head(n)
    coef_low = coef.sort_values(ascending = False).tail(n)
    
    return coef_high, coef_low
```

```python
def visualize_coefficient(models):
    fig, axs = plt.subplots(figsize = (24,10), nrows=1, ncols=3)
    fig.tight_layout()
    
    for i_num, model in enumerate(models):
        coef_high, coef_low = get_top_bottom_coef(model)
        coef_concat = pd.concat([coef_high, coef_low])
        
        axs[i_num].set_title(model.__class__.__name__+'Coefficients', size = 25)
        axs[i_num].tick_params(axis='y', direction='in', pad = -120)
        
        for label in (axs[i_num].get_xticklabels() + axs[i_num].get_yticklabels()):
            label.set_fontsize(22)
        sns.barplot(x=coef_concat.values, y=coef_concat.index, ax=axs[i_num])
        
```

```python
#아래 그래프를 보면 라쏘회귀에서의 튜닝이 필요함이 보인다. 

visualize_coefficient(models)
```

```python
def get_avg_rmse_cv(models):
    for model in models:
        rmse_list = np.sqrt(-cross_val_score(model, X_features, target,
                                            scoring = 'neg_mean_squared_error', cv = 5))
        rmse_avg = np.mean(rmse_list)
        print('\n{0} CV RMSE 값 리스트: {1}'.format(model.__class__.__name__, np.round(rmse_list, 3)))
        print('{0} CV 평균 RMSE 값: {1}'.format(model.__class__.__name__, np.round(rmse_avg, 3)))
```

```python
def print_best_params(model, params):
    grid_model = GridSearchCV(model, param_grid = params, 
                              scoring = 'neg_mean_squared_error', cv = 5)
    grid_model.fit(X_features, target)
    rmse = np.sqrt(-1 * grid_model.best_score_)
    print('{0} 5 CV 시 최적 평균 RMSLE 값: {1}, 최적 alpha: {2}'.format(model.__class__.__name__, 
                                                                np.round(rmse,4), grid_model.best_params_))
```
